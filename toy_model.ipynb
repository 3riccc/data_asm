```python
import numpy as np
import random
import networkx as nx
import math
from tqdm import tqdm
import copy
import matplotlib.pyplot as plt

np.random.seed(2051)
random.seed(2050)
```

```python
N = 20# 有信息投资者数量
T = 100#Number of trading days in simulation
epochs = 1000 # 重新初始化的次数
kesi_gov= 0.5#公共信息波动的方差
g = 0.05 #每次交易拿出自己的本金的百分之几
C1 = 1 #Threshold value of dependence on neighbor information
C2 = 0.1 #Threshold value of dependence on public information
lambda_ = 0.25 # 市场深度



G = nx.random_graphs.watts_strogatz_graph(N, 2, 0.3) # 小世界 ， 每个节点4个邻居，重连概率0.3
# G =  nx.random_graphs.barabasi_albert_graph(N, 1) # 无标度, 每次加入1条边这
# G = nx.random_graphs.erdos_renyi_graph(N, 0.2)  # ER随机图，Nnodes，任意节点连接概率0.2
# G = nx.random_graphs.random_regular_graph(3, N) # 规则图，每个节点3个邻居
net =  nx.to_numpy_matrix(G)
# 数组格式不对，在K[:,i]的时候会出问题，这样解决
temp = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        temp[i,j] = net[i,j]
net = temp
```

```python
net.shape
```

```python
# 一个人的投资决策
def trade_decision(S,rt,n,i,U,t,sigmat,pt_1,cashs,stocks):
    nei_s =  K[:,i] *S # 邻居情绪加权
    nei_info = np.mean(nei_s)  # 邻居最终的信息
        
    
    # 自己的投资决策
    si = 0
    vi = 0
    wi = C1*nei_info + C2*U*n #+ C3[i]*ksi
    if wi >= 0 and wi < 1:
        vi = g*cashs[i]/pt_1*wi
    elif wi >= 1:
        vi = g*cashs[i]/pt_1
    if wi < 0 and wi > -1:
        vi = g * stocks[i] * wi
    elif wi <= -1:
        vi = g * stocks[i] * -1
        
    return vi
# 更新大家的资产
def asset_update(action,pt,cashs,stocks):
    for i in range(N):
        cashs[i] -=pt*action[i]
        stocks[i] +=action[i]
    return cashs,stocks

# 做市商定价
def make_price(S,pt1):
    pt = 10**(np.mean(S)/lambda_+ np.log10(pt_1)) # 随着节点数量的增加，净需求增大，所以平均或许是合理的选择
    return pt

# 市场的一天
def one_day(rt,pt_1,n,action,U,t,sigmat,cashs,stocks):
    # 公共信息更新
    n = n + np.random.normal(0,kesi_gov)
    S2 = np.random.randn(N)
    # 投资决策
    for i in range(N):
        S2[i] = trade_decision(action,rt,n,i,U,t,sigmat,pt_1,cashs,stocks)
    # 价格更新
    p = make_price(S2,pt_1)
    #资产更新
    cashs,stocks = asset_update(S2,p,cashs,stocks)
    return p,n,S2,cashs,stocks
```

### 运行


```python
sample_ns = np.zeros((epochs,T))
sample_price_series = np.zeros((epochs,T))
sample_action = np.zeros((epochs,T,N))
sample_cash = np.zeros((epochs,T,N))
sample_stock = np.zeros((epochs,T,N))
```

```python
# 对邻居的信任程度
K = net
# 对公共信息信任程度
# U = np.random.uniform(0,1)
U = 1
```

```python
e = 0
while e < epochs:
    print(e)
    cashs = 1000*np.ones(N)#所有人的初始现金为1 
    stocks = np.ones(N)#所有人的初始股票为1
    
    ps = np.random.randn(10)+1000
    ps = ps.tolist()
    
    
    n = np.random.normal(0,1)#初始宏观信息
    action = np.random.uniform(-1,1,N)#初始动作
    
    
    #record
    price_series = [ps[-1]]
    
    ns = np.zeros(T)
    one_sample_price = np.zeros(T)
    one_sample_stock = np.zeros((T,N))
    one_sample_cash = np.zeros((T,N))
    actions = np.zeros((T,N))
#     one_sample_cash[e,0,:] = np.array(cashs)
    
    
    for t in range(T):
        rt = np.log10(ps[-1]) - np.log10(ps[-2])# 昨天的收益
        sigmat = 0
        
        # 昨天的信息
        pt_1 = ps[-1]
        
        one_sample_cash[t,:] = np.array(copy.copy(cashs))
        one_sample_stock[t,:] = np.array(copy.copy(stocks))
        actions[t,:] = action
        ns[t] = n
        one_sample_price[t] = pt_1


        p,n,action,cashs,stocks= one_day(rt,pt_1,n,action,U,t,sigmat,cashs,stocks)
        ps.append(p)
        price_series.append(p)
    if np.max(one_sample_price > 10000) or np.min(one_sample_price < 50):
        print('restart')
        continue
    if math.isnan(np.sum(actions)):
        print('nan -> restart')
    else:
        sample_price_series[e] = one_sample_price
        sample_cash[e] = one_sample_cash
        sample_stock[e] = one_sample_stock
        sample_action[e] = actions
        sample_ns[e] = ns
        e += 1
print('epoch,',epochs)
print('node,',N)
print('T',T)
print('action,',sample_action.shape)
print('cash:',sample_cash.shape)
print('price:',sample_price_series.shape)
print('stock:',sample_stock.shape)
print('ns:',sample_ns.shape)
```

```python
for i in range(epochs):
    if max(sample_price_series[i,:] > 1e42):
        print(i)
    plt.plot(sample_price_series[i,:])
plt.title('price')
plt.show()
```
