```python
import numpy as np
import random
import networkx as nx
```

### 初始化

```python
np.random.seed(2050)
random.seed(2050)
N = 500 # 

# 更新情绪时候用的参数，alpha，beta，gama
maxgama = 3
maxalpha = 0.2
absum = 1 # sum of alpha and beta

# 所有人的初始现金
allcash = 360000
# 所有人的初始股票
alla = 4651

# 网络的连接强度
scale=0.3

epoch_num = 1000

# 最开始的价格
start_price = 400
```

### 无标度初始化


#### 初始现金和股票

```python
temp = 0
for i in range(1,N+1):
    temp += i/float(i)
prop = allcash / temp
C = []
for i in range(1,N+1):
    C.append(prop*(1/i)*prop)
```

```python

temp = 0
for i in range(1,N+1):
    temp += i/float(i)
prop = alla / temp
A = []
for i in range(1,N+1):
    A.append(prop*(1/i)*prop)
```

#### T

```python
# 所有人的T
T = []
for i in range(N):
    T.append(int(random.uniform(10,100)))
```

#### 网络结构和权重

```python
# 所有人的影响力，表现为这个人连接到其他任何一个人的概率，
# ith agent is rnadomly connected to a set of other agents whose number and strength is inversely proportional to the rank of the ith agent
pouts = []

for i in range(1,N+1):
    pouts.append((1/float(i))*scale)
sumnum = 0
for i in range(1,N+1):
    sumnum += 1/i

probs = np.zeros((N,N)) #各个位置为1的概率
for i in range(N):
    links = N*pouts[i] #这一行一共有多少个1
    temp = links / sumnum
    for j in range(N):
        probs[i][j] = temp * (1/(j+1))
strengths = []
for i in range(1,N+1):
    strengths.append((1/i)*5)
net = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        if i == j:
            continue
        if random.uniform(0,1) < probs[i][j]:
            net[i][j] = strengths[i]
        else:
            net[i][j] = 0
net01 = np.sign(net)
print('nodes:',N)
print('strengths::',np.sum(net))
print('edges:',np.sum(net01))
```

#### 更新情绪时候的参数

```python


# alpha 反比于节点序号，越往前的节点越不容易被市场影响
alpha = []
for i in range(1,N+1):
    alpha.append(maxalpha * (1/i))
alpha = np.array(alpha)

beta  = absum-alpha
# beta = 0.1 - alpha

# gama 正比于节点的序号，越往后的节点越容易被市场影响
gama = np.zeros(N)
temp = maxgama / N
for i in range(N):
    gama[i] = (i+1)*temp
gama = np.array(gama)

# beta 和 gama均有0.5的可能为正或负
# beta = beta * np.sign(np.random.uniform(-1,1,N))
gama = gama * np.sign(np.random.uniform(-1,1,N))
```

#### 初始情绪

```python
S = np.random.uniform(-1,1,N)
```

### 节点的情绪更新

```python
def F(x):
    return max(min(x,1),-1)
#     return np.tanh(x)

# i，要更新哪个人？
# rt 昨天的回报
def update_emotion(i,rt):
    global alpha
    global beta
    global gama
    global S
    global net
    
    # 邻居的加权平均
    ## 分母应该是连接强度的绝对值求和
    if np.sum(net[:,i] * S) == 0: #如果没有邻居
        nei_s = 0
    else:
        nei_s = np.sum(net[:,i] * S) / (np.sum(net[:,i]))
    
    si = alpha[i] * S[i] + beta[i] * nei_s + gama[i]*rt 
    si = F(si)
    return si,nei_s
```

### 计算价格波动性

```python
# 一段时间内的价格波动性即这段时间内的回报的均值
def volatility(ps):
    a = ps-np.mean(ps)
    b = a**2
    c = np.sum(b) / (a.shape[0]-1)
    return np.sqrt(c)
```

### 节点的投资决策，情绪>0，买

```python
# i，哪个人
# p，昨天的价格
# ps，需要计算波动性的最近一段的价格，每个人不一样
def trade_decision_buy(i,p,ps):
    global C
    global A
    ci= S[i] * C[i] # ci花掉的现金
    ai = ci / p
    p_next = p * (1+S[i])*np.random.normal(1,0.001)
    if random.uniform(0,1) < 0.05: #0.05的可能性参与交易
        return ai,p_next
    else:
        return -1,-1
    
```

### 节点投资决策，情绪<0，卖

```python
# i，哪个人
# p，昨天的价格
# ps，需要计算波动性的最近一段的价格，每个人不一样
def trade_decision_sell(i,p,ps):
    global S
    ai = abs(S[i]) * A[i] # 要卖的股票数量
    p_next = p * (1+S[i]) * np.random.normal(1,0.001)
    if random.uniform(0,1) < 0.05: #0.05的可能性参与交易
        return ai,p_next
    else:
        return -1,-1
```

### 做市商定价

```python
# 计算买的总股数
def buy_amount(bA,bP,p):
    bamount = 0
    for i in range(N):
        if bP[i] >= p and bP[i] != -1:# 如果做市商给的价格  低于  我的预期价格，我就买
            bamount += bA[i] ## 小a
    return bamount
# 计算卖的总股数
def sell_amount(sA,sP,p):
    samount = 0
    for i in range(N):
        if sP[i] <= p and sP[i] != -1:# 如果做市商给的价格  高于  我的预期价格，我就卖
            samount += sA[i]
    return samount

# 获取现在买卖的总数量
def get_total(A):
    tnum = 0
    for i in range(N):
        if A[i] != -1:
            tnum += A[i]
    return tnum

# S : 今天所有人的操作
# bP:买的人的预期价格
# bA：买的人的预期数量
# sA，sP同理
# pt_1 : 昨天的价格
def make_price(bA,bP,sA,sP):
#     print(bA)
#     print(bP)
#     print(sA)
#     print(sP)
    # 所有的价格
    ps = []
    for i in range(N):
        if bP[i] > 0:
            ps.append(bP[i])
        if sP[i] > 0:
            ps.append(sP[i])
    ps = sorted(ps)
#     print('all price:',ps)
    ps1 = np.array(ps[:-1])
    ps2 = np.array(ps[1:])
    ps = 0.5 * (ps2+ps1)

    # 找出出清价格p
    for p in ps:
        b_amount = buy_amount(bA,bP,p)
        s_amount = sell_amount(sA,sP,p)
        if s_amount >= b_amount: # 卖的>=买的，则出清 ，## 改成刚好成交
            print('出清价格：',p)
            return p
    # 如果未出清
    print('total buyer amount,',get_total(bA))
    print('total seller amount',get_total(sA))
    print('未出清')
    return -10000
```

### 更新大家的持股数量和现金数量

```python
# 更新大家的现金
def update_CA(bA,bP,sA,sP,p):
    global A
    global C
    # 看买入的人
    for i in range(N):
        if bP[i] > p and bP[i] != -1: # 成交，按价格P买入
            A[i] += bA[i]
            C[i] -= p*bA[i]  ### 这里不对
    # 看卖出的人
    for i in range(N):
        if sP[i] < p and sP[i] != -1: # 按价格p卖出
            A[i] -= sA[i]
            C[i] += sA[i] * p ### 这里也不对
```

### 市场的一天

```python
# rt 昨天的收益
# pt_1 昨天的价格
# ps，所有历史价格序列
def one_day(rt,pt_1,ps):
    global S
    global NEI
    global emo_one
    # 所有人更新自己的情绪
#     print('emotion before update:')
#     print(S[:20])
    todayS = [0 for i in range(N)]
    for i in range(N):
        si,nei_s = update_emotion(i,rt)
        todayS[i] = si
        if i == 1:
            NEI.append(nei_s)
            emo_one.append(si)
    S = todayS
#     print('emotion after update:')
#     print(S[:20])




    
    # 所有人进行投资决策
    bA = [-1 for i in range(N)] # 买的数量，初始化，-1
    bP = [-1 for i in range(N)] # 买价格
    sA = [-1 for i in range(N)] # 卖数量
    sP = [-1 for i in range(N)] # 卖价格
    for i in range(N):
        if todayS[i] > 0: # 这个人要buy
            Ai,pi = trade_decision_buy(i,pt_1,ps[-T[i]:])
            bA[i] = Ai # 要买的数量
            bP[i] = pi # 买的报价
        else: # 这个人要sell
            Ai,pi = trade_decision_sell(i,pt_1,ps[-T[i]:])
            sA[i] = Ai # 卖的数量
            sP[i] = pi # 卖的报价

    # 做市商更新今天的价格
    p = make_price(bA,bP,sA,sP)

    
    if p == -10000:#表示没有出清
        return p
    
    # 大家进行交易，更新大家的持股数量和钱数
    update_CA(bA,bP,sA,sP,p)
    return p
```

### 市场的运转

```python
ps1 = [start_price]
for i in range(100):
    p = ps1[-1]+np.random.normal(0,0.01)
    ps1.append(p)
# ps = np.load('p.npy').squeeze()
# ps = ps.tolist()

ss = [[] for i in range(N)] # emotion for all
ts = [] # total emotion


NEI = []
emo_one = []
rts = []

for t in range(epoch_num):
    ts.append(np.sum(S))
    # 记录大家的情绪
    for i in range(N):
        ss[i].append(S[i])
    print('day:',t)
    if t% 100 == 0:
        print('current step:',t)
    rt = np.log(ps1[-1]) - np.log(ps1[-2]) # 昨天的收益
    print('rt:',rt)
    pt_1 = ps1[-1] # 昨天的价格
    # 开始今天的交易
    p = one_day(rt,pt_1,ps1)
    # 没有出清
    if p == -10000:
        continue
    
    # 记录今天的价格
    rts.append(rt)
    ps1.append(p)
    print(p)
   
```
