```python
import numpy as np
import random
import networkx as nx
from tqdm import tqdm
import copy
import matplotlib.pyplot as plt
import math

np.random.seed(2051)
random.seed(2051)
```

```python
N = 300# 有信息投资者数量
T = 100#Number of trading days in simulation
epochs = 1000# 重新初始化的次数
# epochs = 10# 重新初始化的次数
kesi_gov= 1#公共信息波动的方差
kesi_pri = 1 #个人信息的方差
g = 0.25 #每次交易拿出自己的本金的百分之几
C1 = 1.6#邻居信息信任
C2 = 0.4 #公共信息信任
C3 = 2 # 私人信息信任

C1 = np.random.uniform(0,C1,N)
C2 = np.random.uniform(0,C2,N)
C3 = np.random.uniform(0,C3,N)

# 简化一下C1，C2，C3，消除抑制性，看看好不好学
# C1 = np.ones(N)*C1
# C2 = np.ones(N)*C2
# C3 = np.ones(N)*C3




W = np.random.uniform(0,0.001,N)

lambda_ = 0.25 # 市场深度



G = nx.random_graphs.watts_strogatz_graph(N, 2, 0.3) # 小世界 ， 每个节点4个邻居，重连概率0.3
# G =  nx.random_graphs.barabasi_albert_graph(N, 1) # 无标度, 每次加入1条边这
# G = nx.random_graphs.erdos_renyi_graph(N, 0.2)  # ER随机图，Nnodes，任意节点连接概率0.2
# G = nx.random_graphs.random_regular_graph(3, N) # 规则图，每个节点3个邻居
net =  nx.to_numpy_matrix(G)
# 数组格式不对，在K[:,i]的时候会出问题，这样解决
temp = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        temp[i,j] = net[i,j]
net = temp
```

```python
def init_cash(allcash):
    temp = 0
    for i in range(1,N+1):
        temp += 1/float(i)
    prop = allcash / temp
    C = []
    for i in range(1,N+1):
        C.append(prop*(1/i))
    return C
def init_stock(alla):
    temp = 0
    for i in range(1,N+1):
        temp += 1/float(i)
    prop = alla / temp
    A = []
    for i in range(1,N+1):
        A.append(prop*(1/i))
    return A
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x) )/ (math.exp(x)+math.exp(-x))
# 一个人的投资决策
def trade_decision(S,rt,n,i,U,t,sigmat,pt_1,cashs,stocks):
    nei_s =  K[:,i] *S # 邻居情绪加权
    nei_info = np.sum(nei_s)  # 邻居最终的信息
    ksi = np.random.normal(0,kesi_pri) #个人信息
        
    
    # 自己的投资决策
    si = 0
    vi = 0
    wi = C1[i]*nei_info + C2[i]*U*n + C3[i]*ksi
#     print(wi)
    if wi >= W[i]:
        vi = g*cashs[i]/pt_1*wi
    if wi < -W[i]:
        vi = g*stocks[i]*wi
    if math.isinf(vi):
        print('got inf')
        print('node:',i)
        print('nei',nei_info)
        print('wi',wi)
        print('stock',stocks[i])
        print('cash',cashs[i])
        print('pt-1',pt_1)
        print(' vi = g*cashs[i]/pt_1*wi')
        
        d()
    if math.isnan(vi):
        print('---sum action')
        print(S)
        print(nei_s)
        print(nei_info)
        
        print('wi')
        print(wi)
        
        print(cashs[i])
        print(pt_1)
        print(stocks[i])
        d()
# #     print(vi)
#     return tanh(vi)
    return max(min(vi,1),-1)
# 更新大家的资产
def asset_update(action,pt,cashs,stocks):
    # 股票和现金数量不变
#     cashs = init_cash(1000*400)
#     stocks = init_stock(400)
    
    for i in range(N):
        # 这个人要买入
        if action[i] > 0:
        # 剩下的钱够买
            if cashs[i] > pt*action[i]:
                cashs[i] -=pt*action[i]
                stocks[i] +=action[i]
            else:
                # 剩下的钱不够买，那么就全款买入
                stocks[i] += cashs[i] / pt
                cashs[i] = 0
        # 这个人要卖出
        else:
            # 这个人有足够的股票卖
            if stocks[i] > -action[i]:
                stocks[i] += action[i]
                cashs[i] -= pt*action[i]
            else:
                # 他要卖的数量大于他持有的数量，就卖空
                cashs[i] += stocks[i] * pt
                stocks[i] = 0
        
    return cashs,stocks

# 做市商定价
def make_price(S,pt1):
    pt = 10**(np.mean(S)/lambda_+ np.log10(pt_1)) # 随着节点数量的增加，净需求增大，所以平均或许是合理的选择
    return pt

# 市场的一天
def one_day(rt,pt_1,n,action,U,t,sigmat,cashs,stocks):
    # 公共信息更新
#     n = n + np.random.normal(0,kesi_gov)


    # 公共信息具有调控作用，当价格过低时，公共信息促使价格升高，反之亦然，价格合理时，公共信息不变
    if pt_1 < 300:
        n = n + np.random.normal(1,kesi_gov)
    elif pt_1 > 10000:
        n = n + np.random.normal(-1,kesi_gov)
    else:
        n = n + np.random.normal(0,kesi_gov)


    S2 = np.random.randn(N)
    
    # 投资决策
    for i in range(N):
        S2[i] = trade_decision(action,rt,n,i,U,t,sigmat,pt_1,cashs,stocks)

    # 价格更新
    p = make_price(S2,pt_1)
    #资产更新
    cashs,stocks = asset_update(S2,p,cashs,stocks)

            
    return p,n,S2,cashs,stocks
```

### 运行


```python
sample_ns = np.zeros((epochs,T))
sample_price_series = np.zeros((epochs,T))
sample_action = np.zeros((epochs,T,N))
sample_cash = np.zeros((epochs,T,N))
sample_stock = np.zeros((epochs,T,N))
```

```python
# 对邻居的信任程度
K = net
# 对公共信息信任程度
# U = np.random.uniform(0,1)
U = 1
```

```python
e=0
while e < epochs:
    print(e)
    cashs = 1000*np.ones(N)#所有人的初始现金为1 
    stocks = np.ones(N)*1#所有人的初始股票为1
#     cashs = init_cash(1000*400)
#     stocks = init_stock(400)
    
    ps = np.random.randn(10)+1000
    ps = ps.tolist()
    
    
    n = np.random.normal(0,1)#初始宏观信息
    action = np.random.uniform(-1,1,N)#初始动作
    
    
    #record
    price_series = [ps[-1]]
    
    ns = np.zeros(T)
    one_sample_price = np.zeros(T)
    one_sample_stock = np.zeros((T,N))
    one_sample_cash = np.zeros((T,N))
    actions = np.zeros((T,N))
#     one_sample_cash[e,0,:] = np.array(cashs)
    
    all_cash = []
    all_stock = []
    for t in range(T):
        
        # record
        all_cash.append(np.sum(cashs))
        all_stock.append(np.sum(stocks))
        if t % 100 == 0:
            print(t)
        
        
        rt = np.log10(ps[-1]) - np.log10(ps[-2])# 昨天的收益
        sigmat = 0
        
        # 昨天的信息
        pt_1 = ps[-1]
        
        one_sample_cash[t,:] = np.array(copy.copy(cashs))
        one_sample_stock[t,:] = np.array(copy.copy(stocks))
        actions[t,:] = action
        ns[t] = n
        one_sample_price[t] = pt_1


        p,n,action,cashs,stocks= one_day(rt,pt_1,n,action,U,t,sigmat,cashs,stocks)
        ps.append(p)
        price_series.append(p)
#     if np.max(one_sample_price > 6000):
#         print('too big => restart')
#         continue
#     if  np.min(one_sample_price) <= 10:
#         print('too small -> restart')
#         continue
    if math.isnan(np.sum(actions)):
        print('nan -> restart')
        continue # some action value is nan
    else:
        sample_price_series[e] = one_sample_price
        sample_cash[e] = one_sample_cash
        sample_stock[e] = one_sample_stock
        sample_action[e] = actions
        sample_ns[e] = ns
        e = e+1

print('epoch,',epochs)
print('node,',N)
print('T',T)
print('action,',sample_action.shape)
print('cash:',sample_cash.shape)
print('price:',sample_price_series.shape)
print('stock:',sample_stock.shape)
print('ns:',sample_ns.shape)
```
