```python
import numpy as np
import networkx as nx
```

```python
net = np.load('/Users/eden/project/research/data/agent_quant/tr_new/net.npy')
```

```python
G = nx.from_numpy_matrix(net)
```

```python
node_labels = G.nodes()
ld = {}#label dict
for no in G.nodes():
    ld[no] = '\n\n'+str(no)
nx.draw(G,pos=nx.spring_layout(G),labels=ld, with_labels = True,font_size=18,node_color='green',alpha=0.6)
```
