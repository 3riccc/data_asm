```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p_series = np.load('./para/price_generated.npy')
```

```python
plt.plot(p_series)
plt.show()
```

```python
p_today = p_series[:-1]
p_tomorrow = p_series[1:]
```

```python
p_diff = p_tomorrow - p_today
plt.plot(p_diff)
plt.show()
```

#### 平均两天的价格差

```python
print(np.mean(np.abs(p_diff)))
```

```python
print(np.mean(np.abs(p_diff[:100])))
```
